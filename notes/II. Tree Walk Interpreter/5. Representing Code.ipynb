## Chomsky Hierarchy 

- classification of formal language based on how complex their grammar rules are
and what kind of machine is needed to recognize them

### 4 Levels:

    - Type 3: Regular (Finite State Machine) ex: identifiers, regex
        - Simplest Class
        - ex: ab*
        - Matches: a, ab, abb, abbb
        - used heavily in lexers/tokenizers, search, text maching
        - cannot handle arbitrary nesting, balanced parantheses

    - Type 2: Context Free (Pushdown Automation) ex: nested parentheses
        - can handle recursive structure
        - pushdown automatata (finite automaton + stack)
        - ex: ((()))
        - Grammar: S-> (S), S → ε
        - most programming languages are mostly context free. 
        - Used in 
            - Parsers
            - Compiler Syntax Analysis
            - AST generation
    

    - Type 1: Context-Sensitive (Linear Bounded Automaton)
        - rules can depend on surrounding context
        - more powerful than Context Free (CFGs)
        - plain text: a^n b^n c^n
        - grammar: abc, aabbcc, aaabbbccc
        - important for advanced language analysis, natural language structures,
        semantic constraints

    - Type 0: Recursively Enumerable (Turing Machine)
        - most general class
        - anything computable by a computer
        - includes all programming languages
