# Scanning

- lexing or scanning or lexical analysis is the first step in any compiler
- takes in raw src code as series of characters and groups it into a series of chunks called tokens
- goal is to have a full featured, fast scanner that can take any string of Lox src code and produce the tokens that will be fed to the parser

## Lexemes and Tokens

- lexemes: each of the blobs of characters. Lexical analysis job is to scan list of characters and group them in smallest sequences to represent something
    - the actual text fragment in source code that the lexer recognizes as a token.
    
- token type: categorize tokens from the raw lexeme by comparing strings. We have different type for each keyword, operator, bit of punctuation, and literal type

## Regular Languages and Expressions

- core of scanner is a loop, scanner figures out what lexeme the character belongs to, and consumes it and any following characters that are part of that lexeme
- lexical grammar
    - rules that determine how a particular language groups characters into lexemes
    - regular language (rationale language) is a formal language that can be defined by a regular expression
