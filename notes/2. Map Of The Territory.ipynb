# Parts Of A Language


- Below are what most compilers do and what a language takes to being executed

### Scanning

- also known as lexing (lexical analysis)
- takes in linear stream of characters and chunks them together into a series of "words"
- each word is called a "token"
    - tokens could be single characters


### Parsing
- where syntax gets a grammar. Ability to compose larger expressions and statements out of smaller parts
- takes flat sequence of tokens and builds a tree structure
- parse tree or abstract syntax tree

### Static Analysis

- binding / resolution
    - for each identifier, find out where the name is defined and wire the 2 together
    - scope - region of src code where a certain name can be used to refer to certain declaration

- statically typed
    - type check

- locations to store semantic insight
    - often it gets stored right back as attributes on the syntax tree itself, extra fields in the nodes that aren't initialized during parsing but get filled in later
    - store data in lookup tables (symbol table). Keys include identifiers (names of variables and declaration)

### Intermediate Representations

- front end of the pipeline is specific to the src language the program is written in
- back end is concerned with the final architecture where the program will run
- middle is where the code will be stored in some intermediate representation

### Optimization
- once the user's program is understood, we optimize it
- constant folding: if some expression evaluates to the exact same value, evaluation is done at compile time and replace the code with its results.

### Code Generation
- converting optimized code to something machines can understand
- primitive assembly like instructions a CPU runs and not src code humans can read
- we are reaching the descent down the backend


Do we generate instructions for a real CPU or a virtual one?
- if we generate real machine code, we get a exe that the OS can load directly onto the chip
- generating code is complex
- compiler is tied to specific architecture

- to get around this, instructions generated virtual machine code (hypothetical, idealized machine). P-code for portable, but called bytecode

### Virtual Machine

- compiler produces bytecode, chips do not speak bytecode and it is your job to translate
- Option 1: you can write a little mini compiler for each target architecture that converts bytecode to native code for that machine. Bytecode is used as an intermediate representation
- Option 2: write a Virtual Machine (VM), program that emulates a hypothetial chip supporting your virtual architecture at runtime. Bytecode in a VM is slow because it is simulated at runtime each time it executes. Tradeoff is you get simplicity and portability.


#### Clarification:
- virtual machine means different things:
    - system virtual machine emulates an entire hardware platform and OS in software
    - allows you to play Windows games on Linux
    - allows cloud providers to give customers experience of controlling their own "server" w/o needing to allocate separate computers 

- language virtual machines and process virtual machines are different

### Runtime
- user's program into a form that is executable
- for compiled machine code, just tell the OS to load the exe and it runs
- if compiled to bytecode, need to start the VM and load the program into that

# Shortcuts and Alternative Routes

- There are alternative path that some compilers/interpreters take

### Single Pass Compilers

- simple compilers perform parsing, analysis, and code generation so that they produce output code directly in the parser, without ever allocating any syntax trees or other IRs
- single pass compilers restrict the design of the language
    - have no intermediate data structure to store global information
    - memory is precious in the olden days and compilers were not able to hold an entire src file in memory, much less the whole program
- syntax directed translation
    - a structured technique for building all at once compilers
    - associate an action with each piece of grammar

### Tree-Walk Interpreters

- to run the program, interpreter traverses the syntax tree one branch and leaf at a time
- this approach is too slow

### Transpilers

- source-to-source compiler or transcompiler
    - writing complete back end for a language can be lots of work
    - instead you can use some exisiting generic IR to target.
    - Web browsers are the "machines" of today, and their "machine code" is JavaScript
    - [List of languages that targets JS bc it is the way to get code running in your browser.](https://github.com/jashkenas/coffeescript/wiki/list-of-languages-that-compile-to-js)

### Just In Time Compilation

- on the end user's machine, when the program is loaded either from source or platform independent bytecode 

# Compilers and Interpreters

What is the difference between a compiler and an interpreter?

- compiling is an implementation technique that involves translating src language to some other. When you generate bytecode or machine code, you are compiling. When you are transpile to another high lvl language you are compiling also.
    - ex: GCC and Clang take C src code and compiles it to machine code
- interpreters take in source code and executes it immediately. It runs programs from source



- for cPython, the code is parsed and converted to an internal bytecode format and executed in a VM. CPython is an interpreter and a compiler.
