# What Is A Lexeme?

- lexeme is basic unit of meaning in a language
- word or set of word forms that are treated as a single vocab item
- in CS, lexeme refers to actual sequene of characters in src code that matches a token. Lexemes are classified into token types such as keyword, identifier, operator, literal, and punctuations


ex:
- run, runs, ran, running → all belong to the same lexeme: RUN
- go, goes, went, gone, going → all belong to the lexeme: GO
